```python
# Movie Theatre Booking System
seats = 350
sold = 0
bookings = 0
rejected = 0

print("Theatre booking (12+ only, max 15 tickets)")
print("Enter 0 for tickets to exit.\n")

while seats > 0:
    t = int(input("Tickets (0 to exit): "))

    if t == 0:
        break

    if t < 1 or t > 15:
        print("REJECTED - Must be 1-15 tickets")
        rejected += 1
        continue

    if t > seats:
        print("REJECTED - Only", seats, "seats left")
        rejected += 1
        continue

    # Check ages
    ok = True
    for i in range(t):
        age = int(input("Age: "))
        if age < 12:
            print("REJECTED - Age under 12")
            ok = False
            break

    if not ok:
        rejected += 1
        continue

    # Booking confirmed
    seats -= t
    sold += t
    bookings += 1
    print("CONFIRMED -", t, "tickets")

    if seats == 0:
        print("\nTheatre full!")
        break

# Final report
print("\n--- REPORT ---")
print("Total bookings:", bookings)
print("Tickets sold:", sold)
print("Rejected bookings:", rejected)
print("Remaining seats:", seats)
```
